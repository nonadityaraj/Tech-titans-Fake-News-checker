```python
#Feature Extraction
import re # regular expression
```

```python
def wordopt(text):
    # Convert to lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)

    # Remove digits
    text = re.sub(r'\d', '', text)

    # Remove newline characters
    text = re.sub(r'\n', ' ', text)

    return text
    
    
```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
x = news['text']
y = news['label']
```

```python
#training and testing
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3)
```

```python
def output_label(n):
    if n==0:
        return "It is a Fake News"
    elif n==1:
        return "It is Genius News"
    
```

```python
def manual_testing(news):
    # Create dataframe from input
    testing_news = {"text": [news]}
    new_def_test = pd.DataFrame(testing_news)
    
    # Apply the same text cleaning you used during training
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    
    # Get the cleaned text (still string)
    new_x_test = new_def_test["text"]
    
    # VERY IMPORTANT: Transform using the SAME fitted vectorizer!
    new_xv_test = vectorization.transform(new_x_test)
    
    # Now predict using the VECTORIZED input
    pred_lr  = LR.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)
    
    # Optional: If you also want Decision Tree (you commented it out)
    # pred_dtc = dtc.predict(new_xv_test)
    
    print("\n\nLR Prediction: "   + output_label(pred_lr[0]))
    print("GBC Prediction: "  + output_label(pred_gbc[0]))
    print("RFC Prediction: "  + output_label(pred_rfc[0]))
    # print("DTC Prediction: "  + output_label(pred_dtc[0]))   # if you uncomment
```
